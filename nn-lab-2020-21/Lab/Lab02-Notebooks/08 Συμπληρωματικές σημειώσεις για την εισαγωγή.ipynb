# Εγκατάσταση πακέτων Python

Υπάρχουν βασικά δύο τρόποι για την εγκατάσταση πακέτων της Python: α) μέσω του λειτουργικού (αν είμαστε τοπικά), β) μέσω των package managers της Python.

## Παράδειγμα

Αν έχουμε μια διανομή arch linux μπορούμε να εγκαταστήσουμε το tensorflow για GPU (cuda) χρησιμοποιώντας απλά τον package manager της διανομής (του linux μας δηλαδή): ```yay python-tensorflow-cuda```

![](https://linuxtips.us/wp-content/uploads/python-pip-logo.png)

Μπορούμε να επιτύχουμε το ίδιο χρησιμοποιώντας τους package managers της Python, τον pip και την anaconda / conda.

- pip: ```pip install --upgrade tensorflow-gpu``` [(1)](https://www.tensorflow.org/install/pip)
- conda: ```conda create -n tf-gpu tensorflow-gpu``` και
```conda activate tf-gpu``` [(2)](https://www.tensorflow.org/install/pip)

Σημειώστε ότι με το conda κάνουμε χρήση και ενός *virtual environment (venv)* με όνομα "tf-gpu".

![](https://docs.conda.io/en/latest/_images/conda_logo.svg)

## Virtual Environments

H Python μας δίνει την δυνατότητα αντί να εγκαθιστούμε πακέτα και βιβλιοθήκες στο επίπεδο συστήματος (system-wide), να φτιάχνουμε ειδικά εικονικά περιβάλλοντα (virtual envirnments) όπου μπορούμε να εγκαταστήσουμε συγκεκριμένο version της Python και συγκεκριμένα versions διαφόρων βιβλιοθηκών. Αυτό είναι πολύ βολικό σε περιπτώσεις όπου για παράδειγμα μια βιβλιοθήκη θέλει Python 3.6 και μια άλλη 3.7: φτιάχνουμε δύο venvs και εντός τους εγκαθιστούμε την Python που χρειάζεται η κάθε βιβλιοθήκη.

Virtual Environments: [1](https://docs.python.org/3/tutorial/venv.html), [2](https://packaging.python.org/guides/installing-using-pip-and-virtual-environments/), [3](https://docs.python-guide.org/dev/virtualenvs/).

Τα virtual environments είναι επίσης πολύ βολικά όταν δουλεύουμε σε έναν server όπου δεν είμαστε διαχειριστές αλλά απλοί χρήστες. Αν χρειαστούμε συγκεκριμένα versions της Python και των πακέτων απλά φτιάχνουμε venvs και τα εγκαθιστούμε.

# Python εκτός notebooks

- Interactive Session: o πιο απλός τρόπος για να τρέξουμε την Python είναι να γράφουμε απλά: ```python``` στη γραμμή εντολών. Αυτό θα ανοίξει μια διαδραστική συνεδρία (interactive session) όπου μπορούμε να εισάγουμε και εκτελούμε μία-μία εντολές Python.

- Script: ο κλασικός τρόπος εκτέλεσης για scripting languages. Γράφουμε τον κώδικά μας σε ένα αρχείο πχ "myscript.py" με ένα editor και το τρέχουμε έτσι: ```python myscript.py```

- IDE: για μεγαλύτερα προγράμματα που μοιράζονται σε πολλά αρχεία και directories αντί για notebook θα χρησιμοποιήσουμε ένα Integrated Development Environnment (IDE). Μπορούμε για παράδειγμα να χρησιμοποιήσουμε το PyCharm των JetBrains που είναι κορυφαίος σχεδιαστής IDE και μας προσφέρει [δωρεάν τα professional versions](https://www.jetbrains.com/community/education/#students) τους αν κάνουμε εγγραφή με ιδρυματικό email. 

![](https://www.jetbrains.com/pycharm/img/screenshots/simpleLook@2x.jpg)

- repl.it: αν δεν έχουμε ή δεν θέλουμε να χρησιμοποιήσουμε δικούς μας υπολογιστικούς πόρους, μπορούμε να δοκιμάσουμε το [repl.it](https://repl.it/). To repl.it είναι ένα cloud-based IDE που δίνει τη δυνατότητα συνεργατικής εργασίας και υποστηρίζει (και) Python. Οι περιορισμοί στο free tier είναι μικροί.
